```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

tasks = [
    {"Code": "P1", "Task": "Planning"},
    {"Code": "D1", "Task": "Design"},
    {"Code": "DEV1", "Task": "Development"},
    {"Code": "T1", "Task": "Testing"},
    {"Code": "DEP1", "Task": "Deployment"},
]

df_codes = pd.DataFrame(tasks)

sns.set(style="whitegrid")

fig, ax = plt.subplots(figsize=(6, 4))

sns.barplot(x=df_codes["Code"], y=[1]*len(df_codes), hue=df_codes["Task"], dodge=False, ax=ax, palette="muted")

ax.set_xlabel("Activity Code")
ax.set_ylabel("")
ax.set_yticks([])
ax.set_title("Activity Codes and Their Tasks")
ax.legend(title="Task", loc="upper right")

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

tasks = [
    {"Code": "P1", "Task": "Planning", "Start": "2024-02-01", "Duration": 5},
    {"Code": "D1", "Task": "Design", "Start": "2024-02-06", "Duration": 7},
    {"Code": "DEV1", "Task": "Development", "Start": "2024-02-13", "Duration": 10},
    {"Code": "T1", "Task": "Testing", "Start": "2024-02-23", "Duration": 6},
    {"Code": "DEP1", "Task": "Deployment", "Start": "2024-02-29", "Duration": 3},
]

df = pd.DataFrame(tasks)
df["Start"] = pd.to_datetime(df["Start"])
df["End"] = df["Start"] + pd.to_timedelta(df["Duration"], unit="D")

sns.set(style="whitegrid")

fig, ax = plt.subplots(figsize=(10, 5))

for i, task in enumerate(df.itertuples()):
    ax.barh(task.Code, (task.End - task.Start).days, left=task.Start, color=sns.color_palette("muted")[i % 6])

ax.set_xlabel("Date")
ax.set_ylabel("Activity Codes")
ax.set_title("Project Gantt Chart with Activity Codes")
ax.xaxis.set_major_locator(plt.MaxNLocator(6))  # Limit number of dates shown

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
